```python
import pandas as pd
```

```python
combustiveis_df = pd.read_excel("ca-2021-02.xlsx")
display(combustiveis_df)
```

```python
# Inserção simples de dados
combustiveis_df['Ativo'] = True

display(combustiveis_df.head())
```

```python
# Criar uma coluna *Obs* preenchida com MELHOR CIDADE quando a coluna 'Municipio' for igual a 'Sao Paulo'
combustiveis_df['Obs'] = ['Melhor Cidade' if Municipio == 'SAO PAULO' else '' for Municipio in combustiveis_df['Municipio']]
display(combustiveis_df.loc[combustiveis_df['Municipio'].isin(['SAO PAULO', 'INDAIATUBA', 'CAMPINAS', 'SALTO']), ['Municipio', 'Obs']])
```

```python
# Preenchimento de coluna conforme condição 
# Preenchimento de coluna 'Valor de Venda - Status' que verifica: 
# Se o valor de venda for maior que R$6,50 = Caro 
# Caso contrário barato 
import numpy as np

combustiveis_df['Valor de Venda - Status'] = np.where(combustiveis_df['Valor de Venda'] > 6.5, 'Caro', 'Barato')
display(combustiveis_df[['Revenda', 'Valor de Venda', 'Valor de Venda - Status']])
```

```python
# Calcular postos por habitante 
num_habitantes_df = pd.read_csv("ibge_num_habitantes_estimado.csv", sep=";")
#Renomear coluna
num_habitantes_df.rename(columns={"Estado":"Estado - Sigla"}, inplace=True)
display(num_habitantes_df)
```

```python
# MERGE de dois dataframes

colunas = ['Municipio', 'Estado - Sigla']
merge_df = combustiveis_df.merge(num_habitantes_df, how="inner", on=colunas)
display(merge_df)
print(merge_df.info())
```

```python
# Eliminar coluna ou linha vazia (nulas)
# 0 = linhas/ rows
# 1 = colunas/ columns
merge_df.dropna(axis=1, inplace=True)
print(merge_df.info())
```

```python
# Eliminar coluna ou linha específica
colunas=['Regiao - Sigla', 'Nome da Rua', 'Numero Rua', 'Bairro', 'Cep', 'Produto', 'Data da Coleta', 'Valor de Venda', 'Unidade de Medida', 'Bandeira', 'Ativo', 'Valor de Venda - Status', 'Obs']
merge_df.drop(labels=colunas, axis=1, inplace=True)
print(merge_df.info())
```

```python
display(merge_df.head(100))
```

```python
# Eliminar linhas duplicadas
merge_df.drop_duplicates(inplace=True)
display(merge_df.head(100))
```

```python
# Agrupar e contar quantos postos têm na cidade
# .count()
postos_por_municipio_df = merge_df.groupby(by=['Estado - Sigla', 'Municipio', 'NumHabitantes2021']).count()
postos_por_municipio_df.reset_index(inplace=True)
# display(postos_por_municipio_df.info())
postos_por_municipio_df.drop('CNPJ da Revenda', axis=1, inplace=True)
postos_por_municipio_df.rename(columns={"Revenda":"NumPostos"}, inplace=True)
display(postos_por_municipio_df)
```

```python
postos_por_municipio_df['NumHabitantesPorPosto'] = postos_por_municipio_df['NumHabitantes2021'] / postos_por_municipio_df['NumPostos']
display(postos_por_municipio_df)
```

```python
# Gráficos
import matplotlib.pyplot as plt

plt.hist(combustiveis_df['Valor de Venda'])

# Título
plt.title("Valor médio dos combustíveis - Nov/2021")

# Rótulo Horizontal
plt.xlabel("Preço (em reais)")
# Rótulo Vertical
plt.ylabel("Quantidade de Coletas")

# Traça a linha vermelha tracejada com o preço médio
# linestyle (dotted = ..) (dashed = --)
plt.axvline(combustiveis_df['Valor de Venda'].mean(), color='red', linestyle='dashed', linewidth=5) 

# "Plota" o gráfico
plt.show()
```

```python
# Visualização do consumo médio
c_mean = combustiveis_df['Valor de Venda'].groupby(by=combustiveis_df['Produto']).mean()
display(c_mean)
```

```python
# Definir a área do gráfico
plt.figure(figsize=(10,5))
#bahr = barras horizontais 
#bar = barras verticais
#c_mean é o dataframe do código anterior
#alpha = opacidade
c_mean.plot(
    kind="barh",
    xlabel="Tipo de Combustível",
    ylabel="Preço em reais do litro",
    title="Média de preços por combustível",
    color="red",
    alpha=0.5
)

# Grade
# plt.grid()

# Remover as bordas do gráfico


# Exibir
plt.show()
```

```python
# Personalizar planilha
from openpyxl import load_workbook
from openpyxl.styles import PatternFill, Font, Color, Alignment

# Abre excel no openpyxl
wb = load_workbook(excel) # wb = workbook

# Escolher planilha usando o sheet name 
ws = wb['Sumário'] #work sheet -> planilha atual de trabalho

# Pintar cabeçalho da tabela 
cinzinha = PatternFill("solid", fgColor="CCCCCC")
coords = ['A1', 'B1']
for coord in coords:
  ws[coord].fill = cinzinha

# Quando o valor de venda for maior ou igual a R$6,50 pinta de vermelho e negrito
MAX_ROW = ws.max_row
num_linha = 2
while (num_linha <= MAX_ROW):
  coord = 'B'+str(num_linha) #coord="B{0}".format(num_linha)
  if ws[coord].value >= 6.5:
    ws[coord].font = Font(bold=True, color="FF0000")
  num_linha = num_linha + 1

# Salvar o Excel
wb.save(excel)
```
